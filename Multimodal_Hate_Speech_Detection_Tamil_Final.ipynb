```python
import re
from indicnlp.normalize.indic_normalize import IndicNormalizerFactory
from nltk.tokenize import word_tokenize
from indicnlp.tokenize.sentence_tokenize import sentence_split
from indic_transliteration import sanscript
from indic_transliteration.sanscript import transliterate

# Initialize Indic normalizer
factory = IndicNormalizerFactory()
normalizer = factory.get_normalizer("ta")  # Tamil normalizer

# Tamil stopwords (extend as needed)
tamil_stopwords = set(["இந்த", "அந்த", "மற்றும்", "ஒரு", "அதை", "அவர்", "இது", "என", "அவள்", "அவன்", "நான்", "நீ"])

# Preprocessing function
def preprocess_tamil_text(text):
    """
    Preprocess Tamil text for NLP tasks.
    Steps:
    1. Normalize Unicode
    2. Remove unwanted characters
    3. Sentence split (optional)
    4. Tokenize
    5. Remove stopwords
    6. Transliterate (optional)
    """
    # 1. Normalize Unicode
    text = normalizer.normalize(text)

    # 2. Remove unwanted characters (punctuation, numbers, etc.)
    # text = re.sub(r"[^ஂஃஅ-ஔக-ஹௐௗா-் ]", "", text)

    # 3. Sentence splitting (optional)
    sentences = sentence_split(text, lang="ta")

    # 4. Tokenize
    tokens = [word for sentence in sentences for word in word_tokenize(sentence)]

    # 5. Remove stopwords
    tokens = [word for word in tokens if word not in tamil_stopwords]

    # 6. Transliterate (optional - useful for certain models)
    # Uncomment if transliteration is required
    tokens = [transliterate(word, sanscript.TAMIL, sanscript.ITRANS) for word in tokens]

    return " ".join(tokens)
```

```python
import torch
from transformers import AutoTokenizer, AutoModel, Wav2Vec2Processor, Wav2Vec2Model
import torchaudio
import numpy as np
import pandas as pd
from torch.utils.data import Dataset

class ImprovedPreprocessing:
    def __init__(self):
        # Initialize text models
        self.text_tokenizer = AutoTokenizer.from_pretrained("google/muril-base-cased")
        self.text_model = AutoModel.from_pretrained("google/muril-base-cased")

        # Initialize audio models
        self.audio_processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base")
        self.audio_model = Wav2Vec2Model.from_pretrained("facebook/wav2vec2-base")

        # Move models to GPU if available
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        self.text_model = self.text_model.to(self.device)
        self.audio_model = self.audio_model.to(self.device)

    def extract_text_features(self, text):
        """Extract features using MuRIL"""
        # Tokenize
        inputs = self.text_tokenizer(
            text,
            padding=True,
            truncation=True,
            max_length=128,
            return_tensors="pt"
        ).to(self.device)

        # Get embeddings
        with torch.no_grad():
            outputs = self.text_model(**inputs)
            # Use CLS token embedding
            features = outputs.last_hidden_state[:, 0, :].cpu().numpy()

        return features

    def extract_audio_features(self, audio_path):
        """Extract features using Wav2Vec 2.0"""
        try:
            # Load audio using torchaudio
            waveform, sr = torchaudio.load(audio_path)

            # Resample if sampling rate is not 16k
            if sr != 16000:
                resample_transform = torchaudio.transforms.Resample(orig_freq=sr, new_freq=16000)
                waveform = resample_transform(waveform)

            # Convert to mono if stereo
            if waveform.shape[0] > 1:
                waveform = torch.mean(waveform, dim=0, keepdim=True)

            # Normalize audio to a fixed length (e.g., 5 seconds)
            target_length = 16000 * 5  # 5 seconds at 16kHz
            if waveform.shape[1] < target_length:
                # Pad with zeros if shorter than target length
                padding = target_length - waveform.shape[1]
                waveform = torch.nn.functional.pad(waveform, (0, padding))
            else:
                # Truncate if longer than target length
                waveform = waveform[:, :target_length]

            # Process the audio using Wav2Vec2Processor
            inputs = self.audio_processor(waveform.squeeze().numpy(), sampling_rate=16000, return_tensors="pt").to(self.device)

            # Get embeddings
            with torch.no_grad():
                outputs = self.audio_model(**inputs)
                # Use mean pooling over time dimension
                features = outputs.last_hidden_state.mean(dim=1).cpu().numpy()

            return features.squeeze()  # Ensure consistent shape

        except Exception as e:
            print(f"Error processing audio file {audio_path}: {str(e)}")
            # Return zeros with the expected shape if processing fails
            return np.zeros((768,))  # Adjust this size based on your model's output dimension

def preprocess_multimodal_data(audio_df, text_df):
    processor = ImprovedPreprocessing()

    # Merge dataframes
    merged_df = pd.merge(audio_df, text_df, on='File Name', how='inner')
    print(f"Total samples after merging: {merged_df.shape[0]}")

    # Extract improved features
    text_features = []
    audio_features = []
    valid_indices = []

    for idx, row in merged_df.iterrows():
        try:
            # Extract text features
            text_feat = processor.extract_text_features(row['Processed_Text'])

            # Extract audio features
            audio_feat = processor.extract_audio_features(row['File Path'])

            # Only append if both features are valid and have the expected shape
            if text_feat.squeeze().shape == (768,) and audio_feat.shape == (768,):
                text_features.append(text_feat.squeeze())
                audio_features.append(audio_feat)
                valid_indices.append(idx)
            else:
                print(f"Skipping sample {idx} due to unexpected feature shape")

            # Print progress
            if idx % 10 == 0:
                print(f"Processed {idx}/{len(merged_df)} samples")

        except Exception as e:
            print(f"Error processing sample {idx}: {str(e)}")
            continue

    # Filter the DataFrame to keep only valid samples
    merged_df = merged_df.iloc[valid_indices]

    # Convert to tensors
    text_features = torch.FloatTensor(np.stack(text_features))
    audio_features = torch.FloatTensor(np.stack(audio_features))
    labels = torch.LongTensor(merged_df['Class Label Short'].astype('category').cat.codes)

    print("Final shapes:")
    print("Audio Features:", audio_features.shape)
    print("Text Features:", text_features.shape)
    print("Labels:", labels.shape)

    return audio_features, text_features, labels

class ImprovedMultiModalDataset(Dataset):
    def __init__(self, audio_features, text_features, labels):
        self.audio_features = audio_features
        self.text_features = text_features
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return {
            'audio': self.audio_features[idx],
            'text': self.text_features[idx],
            'label': self.labels[idx]
        }
```

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
from sklearn.metrics import classification_report, confusion_matrix
from torch.optim import AdamW
from torch.optim.lr_scheduler import CosineAnnealingLR
import numpy as np
from tqdm import tqdm

class MultimodalHateSpeechDetector(nn.Module):
    def __init__(self, num_classes=2):
        super().__init__()

        # Load pre-trained models
        self.text_model = AutoModel.from_pretrained("google/muril-base-cased")
        self.audio_model = Wav2Vec2Model.from_pretrained("facebook/wav2vec2-base")

        # Dimensions
        self.text_dim = self.text_model.config.hidden_size  # Usually 768
        self.audio_dim = self.audio_model.config.hidden_size  # Usually 768

        # Fusion layers
        self.fusion = nn.Sequential(
            nn.Linear(self.text_dim + self.audio_dim, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(256, num_classes)
        )

    def forward(self, text_features, audio_features):
        # Combine features
        combined = torch.cat([text_features, audio_features], dim=1)
        return self.fusion(combined)

class ModelTrainer:
    def __init__(self, model, train_loader, val_loader, device):
        self.model = model.to(device)
        self.train_loader = train_loader
        self.val_loader = val_loader

        self.device = device

        # Training settings
        self.criterion = nn.CrossEntropyLoss()
        self.optimizer = AdamW(model.parameters(), lr=2e-5, weight_decay=0.01)
        self.scheduler = CosineAnnealingLR(self.optimizer, T_max=10)

        # Metrics tracking
        self.best_val_f1 = 0
        self.train_losses = []
        self.val_metrics = []

    def train_epoch(self):
        self.model.train()
        total_loss = 0

        progress_bar = tqdm(self.train_loader, desc='Training')
        for batch in progress_bar:
            # Move batch to device
            audio = batch['audio'].to(self.device)
            text = batch['text'].to(self.device)
            labels = batch['label'].to(self.device)

            # Forward pass
            self.optimizer.zero_grad()
            outputs = self.model(text, audio)
            loss = self.criterion(outputs, labels)

            # Backward pass
            loss.backward()
            self.optimizer.step()

            total_loss += loss.item()
            progress_bar.set_postfix({'loss': loss.item()})

        return total_loss / len(self.train_loader)

    def evaluate(self):
        self.model.eval()
        all_preds = []
        all_labels = []
        val_loss = 0

        with torch.no_grad():
            for batch in tqdm(self.val_loader, desc='Evaluating'):
                # Move batch to device
                audio = batch['audio'].to(self.device)
                text = batch['text'].to(self.device)
                labels = batch['label'].to(self.device)

                # Forward pass
                outputs = self.model(text, audio)
                loss = self.criterion(outputs, labels)
                val_loss += loss.item()

                # Get predictions
                preds = torch.argmax(outputs, dim=1)
                all_preds.extend(preds.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())

        # Calculate metrics
        val_loss = val_loss / len(self.val_loader)
        report = classification_report(all_labels, all_preds, output_dict=True)

        return {
            'loss': val_loss,
            'f1': report['macro avg']['f1-score'],
            'precision': report['macro avg']['precision'],
            'recall': report['macro avg']['recall'],
            'report': report
        }

    def train(self, num_epochs):
        for epoch in range(num_epochs):
            print(f"\nEpoch {epoch+1}/{num_epochs}")

            # Training
            train_loss = self.train_epoch()
            self.train_losses.append(train_loss)

            # Validation
            val_metrics = self.evaluate()
            self.val_metrics.append(val_metrics)

            # Print metrics
            print(f"Train Loss: {train_loss:.4f}")
            print(f"Val Loss: {val_metrics['loss']:.4f}")
            print(f"Val F1: {val_metrics['f1']:.4f}")
            print(f"Val Precision: {val_metrics['precision']:.4f}")
            print(f"Val Recall: {val_metrics['recall']:.4f}")

            # Save best model
            if val_metrics['f1'] > self.best_val_f1:
                self.best_val_f1 = val_metrics['f1']
                torch.save(self.model.state_dict(), 'best_model.pt')

            # Update learning rate
            self.scheduler.step()

# Usage example
def train_hate_speech_detector(train_dataset, val_dataset, batch_size=16, num_epochs=10):
    # Create data loaders
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size)

    # Initialize model
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    # device = torch.device('cpu')  # Switch to CPU for debugging


    model = MultimodalHateSpeechDetector(num_classes=5)
    model.to(device)
    # Initialize trainer
    trainer = ModelTrainer(model, train_loader, val_loader, device)

    # Train model
    trainer.train(num_epochs)

    return model, trainer
```

```python

def get_tamil_synonyms(word, top_n=3):
    """
    Get Tamil synonyms using FastText embeddings.
    """
    try:
        similar_words = fasttext_model.most_similar(word, topn=top_n)
        return [w for w, _ in similar_words]
    except KeyError:
        return []  # Return empty if the word is not in vocabulary

def synonym_replacement(text, n=1):
    """
    Replace 'n' random words in the Tamil text with their synonyms.
    """
    words = indic_tokenize.trivial_tokenize(text)  # Tokenize Tamil text
    new_words = words.copy()

    # Find words that have synonyms
    words_with_synonyms = [word for word in words if get_tamil_synonyms(word)]

    if not words_with_synonyms:
        return text  # No synonyms found, return original text

    num_replacements = min(n, len(words_with_synonyms))
    random.shuffle(words_with_synonyms)
    words_to_replace = words_with_synonyms[:num_replacements]

    for word in words_to_replace:
        synonyms = get_tamil_synonyms(word)
        if synonyms:
            new_words[new_words.index(word)] = random.choice(synonyms)

    return ' '.join(new_words)

# Example Tamil sentence
tamil_text = "நாம் அழகா அசிங்கமா அப்படிங்கறது முக்கியம் கிடையாது லுங்கறது மேட்டரே கிடையாது நம்ம எவ்வளவு அசிங்கமா இருந்தாலும் உன் மைண்டுக்கு உன் மேல ஆசை இருந்துச்சுன்னா உன்ன பார்க்கிறவன் எல்லாம் உன்னை பார்த்து அட்ராக்ட் ஆகலாம்"

# Apply augmentation
augmented_text = synonym_replacement(tamil_text, n=2)
print("Original:", tamil_text)
print("Augmented:", augmented_text)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from sklearn.metrics import classification_report
from transformers import Wav2Vec2ForSequenceClassification, Wav2Vec2Processor
from torch.utils.data import DataLoader, Dataset
import librosa

class AudioDataset(Dataset):
    def __init__(self, audio_paths, labels, processor, max_length=160000):  # 10 seconds at 16kHz
        self.audio_paths = audio_paths
        self.labels = labels
        self.processor = processor
        self.max_length = max_length

    def __len__(self):
        return len(self.audio_paths)

    def preprocess_audio(self, audio_path):
        # Load and resample audio to 16kHz
        try:
            waveform, sr = librosa.load(audio_path, sr=16000)

            # Handle length
            if len(waveform) > self.max_length:
                waveform = waveform[:self.max_length]
            else:
                # Pad with zeros if audio is shorter than max_length
                padding_length = self.max_length - len(waveform)
                waveform = np.pad(waveform, (0, padding_length), 'constant')

            return waveform

        except Exception as e:
            print(f"Error preprocessing audio file {audio_path}: {str(e)}")
            # Return zeros if there's an error
            return np.zeros(self.max_length)

    def __getitem__(self, idx):
        audio = self.preprocess_audio(self.audio_paths[idx])
        label = self.labels[idx]
        return {
            'audio': audio,
            'label': label
        }

def collate_fn(batch):
    """
    Custom collate function to handle preprocessed audio inputs
    """
    audio_inputs = [x['audio'] for x in batch]
    labels = torch.tensor([x['label'] for x in batch])

    # Stack audio inputs into a single numpy array
    audio_array = np.stack(audio_inputs)

    # Process with the Wav2Vec2 processor
    inputs = processor(
        audio_array,
        sampling_rate=16000,
        padding=True,
        return_tensors="pt"
    )

    return {
        'input_values': inputs.input_values,
        'attention_mask': inputs.attention_mask,
        'labels': labels
    }

def train_epoch(model, dataloader, optimizer, criterion, device, gradient_accumulation_steps=4):
    model.train()
    total_loss = 0
    num_batches = 0
    optimizer.zero_grad()

    for i, batch in enumerate(dataloader):
        try:
            input_values = batch['input_values'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            outputs = model(
                input_values=input_values,
                attention_mask=attention_mask
            )

            loss = criterion(outputs.logits, labels)
            loss = loss / gradient_accumulation_steps  # Scale loss
            loss.backward()

            if (i + 1) % gradient_accumulation_steps == 0:
                optimizer.step()
                optimizer.zero_grad()

            total_loss += loss.item() * gradient_accumulation_steps
            num_batches += 1

            # Print progress
            if num_batches % 10 == 0:
                print(f"Processed {num_batches} batches. Current loss: {loss.item():.4f}")

        except Exception as e:
            print(f"Error in training batch: {str(e)}")
            continue

    return total_loss / num_batches if num_batches > 0 else float('inf')

def evaluate(model, dataloader, device):
    model.eval()
    true_labels = []
    predictions = []

    with torch.no_grad():
        for batch in dataloader:
            try:
                input_values = batch['input_values'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                labels = batch['labels']

                outputs = model(
                    input_values=input_values,
                    attention_mask=attention_mask
                )

                _, predicted = torch.max(outputs.logits, dim=1)

                true_labels.extend(labels.cpu().numpy())
                predictions.extend(predicted.cpu().numpy())

            except Exception as e:
                print(f"Error in evaluation batch: {str(e)}")
                continue

    return true_labels, predictions

# Initialize model and processor
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base")
model = Wav2Vec2ForSequenceClassification.from_pretrained(
    "facebook/wav2vec2-base",
    num_labels=5,
    ignore_mismatched_sizes=True
)
model.to(device)

# Initialize training components
criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(model.parameters(), lr=1e-5)  # Reduced learning rate

# Example usage:

# Prepare your data
audio_paths = [...]  # List of paths to your audio files
labels = [...]       # List of corresponding labels

# Create dataset and dataloader
audio_dataset = AudioDataset(audio_paths, labels, processor)
audio_loader = DataLoader(
    audio_dataset,
    batch_size=4,    # Reduced batch size
    collate_fn=collate_fn,
    shuffle=True,
    num_workers=2
)

# Training loop
num_epochs = 3
for epoch in range(num_epochs):
    try:
        print(f"\nEpoch {epoch + 1}/{num_epochs}")
        avg_loss = train_epoch(model, audio_loader, optimizer, criterion, device)
        print(f"Average Loss: {avg_loss:.4f}")

        # Evaluate after each epoch
        true_labels, predictions = evaluate(model, audio_loader, device)
        print("\nClassification Report:")
        print(classification_report(
            true_labels,
            predictions,
            target_names=["Class 0", "Class 1", "Class 2", "Class 3", "Class 4"]
        ))

        # Save checkpoint
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'loss': avg_loss,
        }, f'checkpoint_epoch_{epoch+1}.pt')

    except Exception as e:
        print(f"Error in epoch {epoch + 1}: {str(e)}")
        continue
```

```python
import torch
import torchaudio
import numpy as np
import pandas as pd
from torch.utils.data import Dataset, DataLoader
from transformers import Wav2Vec2Processor, Wav2Vec2Model
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import classification_report

class AudioPreprocessor:
    def __init__(self):
        # Initialize audio models
        self.audio_processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base")
        self.audio_model = Wav2Vec2Model.from_pretrained("facebook/wav2vec2-base")

        # Move model to GPU if available
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        self.audio_model = self.audio_model.to(self.device)

    def extract_audio_features(self, audio_path):
        """Extract features using Wav2Vec 2.0"""
        try:
            # Load audio using torchaudio
            waveform, sr = torchaudio.load(audio_path)

            # Resample if sampling rate is not 16k
            if sr != 16000:
                resample_transform = torchaudio.transforms.Resample(orig_freq=sr, new_freq=16000)
                waveform = resample_transform(waveform)

            # Convert to mono if stereo
            if waveform.shape[0] > 1:
                waveform = torch.mean(waveform, dim=0, keepdim=True)

            # Normalize audio to a fixed length (e.g., 5 seconds)
            target_length = 16000 * 5  # 5 seconds at 16kHz
            if waveform.shape[1] < target_length:
                # Pad with zeros if shorter than target length
                padding = target_length - waveform.shape[1]
                waveform = torch.nn.functional.pad(waveform, (0, padding))
            else:
                # Truncate if longer than target length
                waveform = waveform[:, :target_length]

            # Process the audio using Wav2Vec2Processor
            inputs = self.audio_processor(waveform.squeeze().numpy(),
                                       sampling_rate=16000,
                                       return_tensors="pt").to(self.device)

            # Get embeddings
            with torch.no_grad():
                outputs = self.audio_model(**inputs)
                # Use mean pooling over time dimension
                features = outputs.last_hidden_state.mean(dim=1).cpu().numpy()

            return features.squeeze()  # Ensure consistent shape

        except Exception as e:
            print(f"Error processing audio file {audio_path}: {str(e)}")
            # Return zeros with the expected shape if processing fails
            return np.zeros((768,))  # Wav2Vec2 base model output dimension

def preprocess_audio_data(audio_df):
    processor = AudioPreprocessor()

    audio_features = []
    valid_indices = []

    for idx, row in audio_df.iterrows():
        try:
            # Extract audio features
            audio_feat = processor.extract_audio_features(row['File Path'])

            # Only append if features have the expected shape
            if audio_feat.shape == (768,):
                audio_features.append(audio_feat)
                valid_indices.append(idx)
            else:
                print(f"Skipping sample {idx} due to unexpected feature shape")

            # Print progress
            if idx % 10 == 0:
                print(f"Processed {idx}/{len(audio_df)} samples")

        except Exception as e:
            print(f"Error processing sample {idx}: {str(e)}")
            continue

    # Filter the DataFrame to keep only valid samples
    audio_df = audio_df.iloc[valid_indices]

    # Convert to tensor
    audio_features = torch.FloatTensor(np.stack(audio_features))
    labels = torch.LongTensor(audio_df['Class Label Short'].astype('category').cat.codes)

    print("Final shapes:")
    print("Audio Features:", audio_features.shape)
    print("Labels:", labels.shape)

    return audio_features, labels

class AudioDataset(Dataset):
    def __init__(self, audio_features, labels):
        self.audio_features = audio_features
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return {
            'audio': self.audio_features[idx],
            'label': self.labels[idx]
        }

class AudioClassifier(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.classifier = nn.Sequential(
            nn.Linear(768, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        return self.classifier(x)

# Example usage:

# Preprocess the data
audio_features, labels = preprocess_audio_data(audio_df)

# Create dataset and dataloader
dataset = AudioDataset(audio_features, labels)
train_loader = DataLoader(dataset, batch_size=32, shuffle=True)

```

```python
import torch
import numpy as np
from sklearn.metrics import classification_report

# Assuming you have a test dataset for both text and audio in test_audio_df and test_text_df

# 1. Prepare text and audio datasets
text_test_files = merged_df["Transcript"].tolist()  # Text data (e.g., transcripts)
text_test_labels = merged_df["Class Label Short_x"].map(label_map).tolist()

audio_test_files = merged_df["File Path"].tolist()  # Audio data (e.g., .wav files)
audio_test_labels = merged_df["Class Label Short_x"].map(label_map).tolist()

# Create Datasets and Dataloaders for both text and audio
text_test_dataset = TextDataset(text_test_files, text_test_labels, tokenizer)
audio_test_dataset = AudioDataset(audio_test_files, audio_test_labels, processor)

text_test_loader = DataLoader(text_test_dataset, batch_size=8, shuffle=False)
audio_test_loader = DataLoader(audio_test_dataset, batch_size=8, shuffle=False)

# 2. Evaluate Models on Text and Audio Data

# For Text Model
model_text.eval()
text_preds, text_true_labels = [], []

with torch.no_grad():
    for batch in text_test_loader:
        input_ids, attention_mask, labels = (
            batch["input_ids"].to(device),
            batch["attention_mask"].to(device),
            batch["label"].to(device),
        )
        outputs = model_text(input_ids, attention_mask=attention_mask)
        predictions = torch.argmax(outputs.logits, dim=-1)
        text_preds.extend(predictions.cpu().numpy())
        text_true_labels.extend(labels.cpu().numpy())

# For Audio Model
model_audio.eval()
audio_preds, audio_true_labels = [], []

with torch.no_grad():
    for batch in audio_test_loader:
        input_values, labels = batch["input_values"].to(device), batch["label"].to(device)
        outputs = model_audio(input_values)
        predictions = torch.argmax(outputs.logits, dim=-1)
        audio_preds.extend(predictions.cpu().numpy())
        audio_true_labels.extend(labels.cpu().numpy())

# 3. Late Fusion: Combine the predictions from both models

# One common approach is to use **weighted average** of the logits (or probabilities) for each model
# If models output probabilities, you can combine them; here we assume you have the logits from each model.

# Get the logits (before softmax) from both models for fusion
text_logits = []
audio_logits = []

# Get logits for text model
model_text.eval()
with torch.no_grad():
    for batch in text_test_loader:
        input_ids, attention_mask = (
            batch["input_ids"].to(device),
            batch["attention_mask"].to(device),
        )
        outputs = model_text(input_ids, attention_mask=attention_mask)
        text_logits.extend(outputs.logits.cpu().numpy())

# Get logits for audio model
model_audio.eval()
with torch.no_grad():
    for batch in audio_test_loader:
        input_values = batch["input_values"].to(device)
        outputs = model_audio(input_values)
        audio_logits.extend(outputs.logits.cpu().numpy())

# Convert logits to numpy arrays
text_logits = np.array(text_logits)
audio_logits = np.array(audio_logits)


```

```python
# Fusion: Average the logits (can be weighted as well)
# If the models output logits, you can combine the logits before applying softmax for final prediction
# You could also weight the models based on their importance (e.g., by accuracy on validation data)
fusion_logits = (0.9*text_logits + 0.1*audio_logits) / 2  # Simple average, modify if you want weighted fusion

# Get final predictions by applying softmax
final_preds = np.argmax(fusion_logits, axis=-1)

# 4. Evaluate the final performance
print(classification_report(audio_true_labels, final_preds, target_names=[str(k) for k in label_map.keys()]))
```
