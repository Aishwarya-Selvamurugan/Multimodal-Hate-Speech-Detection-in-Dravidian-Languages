```python
from indicnlp.normalize.indic_normalize import IndicNormalizerFactory
from nltk.tokenize import word_tokenize
from indicnlp.tokenize.sentence_tokenize import sentence_split
from indic_transliteration import sanscript
from indic_transliteration.sanscript import transliterate
import re

# Initialize Indic normalizer for Telugu
factory = IndicNormalizerFactory()
normalizer = factory.get_normalizer("te")  # Telugu normalizer

# Telugu stopwords (extend as needed)
telugu_stopwords = set(["ఈ", "అది", "మరియు", "ఒక", "అటు", "ఇది", "నాకు", "మీకు", "నేను", "నువ్వు", "ఆమె", "ఆయన", "మా", "మీ"])

# Preprocessing function for Telugu
def preprocess_telugu_text(text):
    """
    Preprocess Telugu text for NLP tasks.
    Steps:
    1. Normalize Unicode
    2. Remove unwanted characters
    3. Sentence split (optional)
    4. Tokenize
    5. Remove stopwords
    6. Transliterate (optional)
    """
    # 1. Normalize Unicode
    text = normalizer.normalize(text)

    # 2. Remove unwanted characters (punctuation, numbers, etc.)
    text = re.sub(r"[^ఁ-ఔక-హఽఽా-ౌ్ౠౡౢౣ౦-౯ ]", "", text)

    # 3. Sentence splitting (optional)
    sentences = sentence_split(text, lang="te")

    # 4. Tokenize
    tokens = [word for sentence in sentences for word in word_tokenize(sentence)]

    # 5. Remove stopwords
    tokens = [word for word in tokens if word not in telugu_stopwords]

    # 6. Transliterate (optional - useful for certain models)
    # Uncomment if transliteration is required
    # tokens = [transliterate(word, sanscript.TELUGU, sanscript.ITRANS) for word in tokens]

    return " ".join(tokens)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.utils.class_weight import compute_class_weight
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.utils.data import DataLoader, Dataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import pandas as pd
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
# Load Dataset
df = text_df  # Assuming text_df is your dataset

# Encode Labels
label_map = {label: idx for idx, label in enumerate(df["Class Label Short"].unique())}
df["label"] = df["Class Label Short"].map(label_map)

# Train-Test Split
train_texts, test_texts, train_labels, test_labels = train_test_split(
    df["Transcript"].tolist(), df["label"].tolist(), test_size=0.2, random_state=42
)

# Calculate class weights
class_weights = compute_class_weight('balanced', classes=np.array(list(label_map.values())), y=train_labels)
class_weights = torch.tensor(class_weights, dtype=torch.float).to(device)

# Load Tokenizer
tokenizer = AutoTokenizer.from_pretrained("google/muril-base-cased")

# Custom Dataset Class
class TextDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_length=128):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encoding = self.tokenizer(
            self.texts[idx],
            truncation=True,
            padding="max_length",
            max_length=self.max_length,
            return_tensors="pt",
        )
        return {
            "input_ids": encoding["input_ids"].squeeze(0),
            "attention_mask": encoding["attention_mask"].squeeze(0),
            "label": torch.tensor(self.labels[idx], dtype=torch.long),
        }

# Create Dataset Objects
train_dataset = TextDataset(train_texts, train_labels, tokenizer)
test_dataset = TextDataset(test_texts, test_labels, tokenizer)

# Create Dataloaders
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)

# Load Model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_text = AutoModelForSequenceClassification.from_pretrained(
    "google/muril-base-cased", num_labels=len(label_map)
).to(device)

# Loss & Optimizer with Class Weights
criterion = nn.CrossEntropyLoss(weight=class_weights)
optimizer = optim.AdamW(model_text.parameters(), lr=2e-5)

# Training Loop
for epoch in range(10):  # Adjust the number of epochs as needed
    model_text.train()
    total_loss = 0
    for batch in train_loader:
        input_ids, attention_mask, labels = (
            batch["input_ids"].to(device),
            batch["attention_mask"].to(device),
            batch["label"].to(device),
        )

        optimizer.zero_grad()
        outputs = model_text(input_ids, attention_mask=attention_mask)
        loss = criterion(outputs.logits, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    print(f"Epoch {epoch+1}: Loss = {total_loss / len(train_loader)}")

# Evaluate Model
model_text.eval()
preds, true_labels = [], []

with torch.no_grad():
    for batch in test_loader:
        input_ids, attention_mask, labels = (
            batch["input_ids"].to(device),
            batch["attention_mask"].to(device),
            batch["label"].to(device),
        )

        outputs = model_text(input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        predictions = torch.argmax(logits, dim=-1)
        preds.extend(predictions.cpu().numpy())
        true_labels.extend(labels.cpu().numpy())

# Classification Report
print(classification_report(true_labels, preds, target_names=[str(k) for k in label_map.keys()]))
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.utils.class_weight import compute_class_weight
from transformers import Wav2Vec2Processor, Wav2Vec2ForSequenceClassification
from torch.utils.data import DataLoader, Dataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import librosa
import numpy as np

# Load Dataset (Assume you have a dataframe with 'File Path' and 'Class Label Short' columns)
df = audio_df  # Replace with your actual dataframe containing file paths and labels

# Encode Labels
label_map = {label: idx for idx, label in enumerate(df["Class Label Short"].unique())}
df["label"] = df["Class Label Short"].map(label_map)

# Train-Test Split
train_files, test_files, train_labels, test_labels = train_test_split(
    df["File Path"].tolist(), df["label"].tolist(), test_size=0.2, random_state=42
)

# Calculate Class Weights
class_weights = compute_class_weight('balanced', classes=np.array(list(label_map.values())), y=train_labels)
class_weights = torch.tensor(class_weights, dtype=torch.float)

# Load Wav2Vec2 Processor
processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base")

# Custom Audio Dataset Class
class AudioDataset(Dataset):
    def __init__(self, file_paths, labels, processor, max_length=16000):
        self.file_paths = file_paths
        self.labels = labels
        self.processor = processor
        self.max_length = max_length

    def __len__(self):
        return len(self.file_paths)

    def __getitem__(self, idx):
        file_path = self.file_paths[idx]
        label = self.labels[idx]

        # Load and preprocess audio
        audio, sr = librosa.load(file_path, sr=16000)  # Ensure 16kHz sample rate
        audio = librosa.util.fix_length(audio, size=self.max_length)  # Padding

        # Process with Wav2Vec2
        inputs = self.processor(audio, sampling_rate=16000, return_tensors="pt", padding=True)

        return {
            "input_values": inputs["input_values"].squeeze(0),
            "label": torch.tensor(label, dtype=torch.long),
        }

# Create Dataset Objects
train_dataset = AudioDataset(train_files, train_labels, processor)
test_dataset = AudioDataset(test_files, test_labels, processor)

# Create DataLoaders
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=8, shuffle=False)

# Load Wav2Vec2 Model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_audio = Wav2Vec2ForSequenceClassification.from_pretrained(
    "facebook/wav2vec2-base", num_labels=len(label_map)
).to(device)

# Loss & Optimizer with Class Weights
criterion = nn.CrossEntropyLoss(weight=class_weights.to(device))  # Applying class weights to loss
optimizer = optim.AdamW(model_audio.parameters(), lr=2e-5)

# Training Loop
for epoch in range(10):  # Adjust epochs as needed
    model_audio.train()
    total_loss = 0
    for batch in train_loader:
        input_values, labels = batch["input_values"].to(device), batch["label"].to(device)

        optimizer.zero_grad()
        outputs = model_audio(input_values)
        loss = criterion(outputs.logits, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    print(f"Epoch {epoch+1}: Loss = {total_loss / len(train_loader)}")

# Evaluate Model
model_audio.eval()
preds, true_labels = [], []

with torch.no_grad():
    for batch in test_loader:
        input_values, labels = batch["input_values"].to(device), batch["label"].to(device)

        outputs = model_audio(input_values)
        predictions = torch.argmax(outputs.logits, dim=-1)
        preds.extend(predictions.cpu().numpy())
        true_labels.extend(labels.cpu().numpy())

# Classification Report
print(classification_report(true_labels, preds, target_names=[str(k) for k in label_map.keys()]))
```

```python
import torch
import numpy as np
from sklearn.metrics import classification_report

# Assuming you have a test dataset for both text and audio in test_audio_df and test_text_df

# 1. Prepare text and audio datasets
text_test_files = merged_df["Transcript"].tolist()  # Text data (e.g., transcripts)
text_test_labels = merged_df["Class Label Short_x"].map(label_map).tolist()

audio_test_files = merged_df["File Path"].tolist()  # Audio data (e.g., .wav files)
audio_test_labels = merged_df["Class Label Short_x"].map(label_map).tolist()

# Create Datasets and Dataloaders for both text and audio
text_test_dataset = TextDataset(text_test_files, text_test_labels, tokenizer)
audio_test_dataset = AudioDataset(audio_test_files, audio_test_labels, processor)

text_test_loader = DataLoader(text_test_dataset, batch_size=8, shuffle=False)
audio_test_loader = DataLoader(audio_test_dataset, batch_size=8, shuffle=False)

# 2. Evaluate Models on Text and Audio Data

# For Text Model
model_text.eval()
text_preds, text_true_labels = [], []

with torch.no_grad():
    for batch in text_test_loader:
        input_ids, attention_mask, labels = (
            batch["input_ids"].to(device),
            batch["attention_mask"].to(device),
            batch["label"].to(device),
        )
        outputs = model_text(input_ids, attention_mask=attention_mask)
        predictions = torch.argmax(outputs.logits, dim=-1)
        text_preds.extend(predictions.cpu().numpy())
        text_true_labels.extend(labels.cpu().numpy())

# For Audio Model
model_audio.eval()
audio_preds, audio_true_labels = [], []

with torch.no_grad():
    for batch in audio_test_loader:
        input_values, labels = batch["input_values"].to(device), batch["label"].to(device)
        outputs = model_audio(input_values)
        predictions = torch.argmax(outputs.logits, dim=-1)
        audio_preds.extend(predictions.cpu().numpy())
        audio_true_labels.extend(labels.cpu().numpy())

# 3. Late Fusion: Combine the predictions from both models

# One common approach is to use **weighted average** of the logits (or probabilities) for each model
# If models output probabilities, you can combine them; here we assume you have the logits from each model.

# Get the logits (before softmax) from both models for fusion
text_logits = []
audio_logits = []

# Get logits for text model
model_text.eval()
with torch.no_grad():
    for batch in text_test_loader:
        input_ids, attention_mask = (
            batch["input_ids"].to(device),
            batch["attention_mask"].to(device),
        )
        outputs = model_text(input_ids, attention_mask=attention_mask)
        text_logits.extend(outputs.logits.cpu().numpy())

# Get logits for audio model
model_audio.eval()
with torch.no_grad():
    for batch in audio_test_loader:
        input_values = batch["input_values"].to(device)
        outputs = model_audio(input_values)
        audio_logits.extend(outputs.logits.cpu().numpy())

# Convert logits to numpy arrays
text_logits = np.array(text_logits)
audio_logits = np.array(audio_logits)


```

```python
# Fusion: Average the logits (can be weighted as well)
# If the models output logits, you can combine the logits before applying softmax for final prediction
# You could also weight the models based on their importance (e.g., by accuracy on validation data)
fusion_logits = (0.9*text_logits + 0.1*audio_logits) / 2  # Simple average, modify if you want weighted fusion

# Get final predictions by applying softmax
final_preds = np.argmax(fusion_logits, axis=-1)

# 4. Evaluate the final performance
print(classification_report(audio_true_labels, final_preds, target_names=[str(k) for k in label_map.keys()]))
```
